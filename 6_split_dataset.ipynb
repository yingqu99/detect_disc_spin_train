```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

# path

```python
data_folder = "/home/subblue-3/nextGen/data/attitude_estimation_in_lab_20250820/"
```

# load data

```python
df_roll_all = pd.read_excel(data_folder + "5_roll_high_corr__all_data.xlsx")
df_roll_all.shape
```

```python
df_roll_all
```

```python
df_pitch_all = pd.read_excel(data_folder + "5_pitch_high_corr__all_data.xlsx")
df_pitch_all.shape
```

```python
df_pitch_all
```

# 'true' value (int), others round(2)

```python
df_roll_all_temp = df_roll_all.copy()

for col in df_roll_all_temp.columns:
    if col == 'roll_true':
        df_roll_all_temp[col] = df_roll_all_temp[col].round(2)
    else:
        df_roll_all_temp[col] = df_roll_all_temp[col].astype(int)
        
df_roll_all_temp
```

```python
df_roll_all = df_roll_all_temp
```

```python
df_pitch_all_temp = df_pitch_all.copy()

for col in df_pitch_all_temp.columns:
    if col == 'pitch_true':
        df_pitch_all_temp[col] = df_pitch_all_temp[col].round(2)
    else:
        df_pitch_all_temp[col] = df_pitch_all_temp[col].astype(int)
        
df_pitch_all_temp
```

```python
df_pitch_all = df_pitch_all_temp
```

# split dataset for train, val, test

## X, y

```python
# Set random seed for reproducibility
np.random.seed(42)
```

```python
X_roll = df_roll_all.iloc[:, 1:]
y_roll = df_roll_all.iloc[:, 0]

X_pitch = df_pitch_all.iloc[:, 1:]
y_pitch = df_pitch_all.iloc[:, 0]
```

```python
X_roll
```

## 80% train+val, 20% test

```python
X_roll_temp, X_roll_test, y_roll_temp, y_roll_test = train_test_split(X_roll, y_roll, test_size=0.2, random_state=42, shuffle=True)

X_roll_test.shape, y_roll_test.shape, X_roll_temp.shape, y_roll_temp.shape
```

```python
X_roll_test
```

```python
X_pitch_temp, X_pitch_test, y_pitch_temp, y_pitch_test = train_test_split(X_pitch, y_pitch, test_size=0.2, random_state=42, shuffle=True)

X_pitch_test.shape, y_pitch_test.shape, X_pitch_temp.shape, y_pitch_temp.shape
```

## 85% train, 15% val

```python
X_roll_train, X_roll_val, y_roll_train, y_roll_val = train_test_split(X_roll_temp, y_roll_temp, test_size=0.15, random_state=42, shuffle=True)

X_roll_train.shape, X_roll_val.shape, y_roll_train.shape, y_roll_val.shape
```

```python
X_pitch_train, X_pitch_val, y_pitch_train, y_pitch_val = train_test_split(X_pitch_temp, y_pitch_temp, test_size=0.15, random_state=42, shuffle=True)

X_pitch_train.shape, X_pitch_val.shape, y_pitch_train.shape, y_pitch_val.shape
```

## create df

```python
df_roll_train = pd.concat((y_roll_train, X_roll_train), axis=1)
df_roll_val = pd.concat((y_roll_val, X_roll_val), axis=1)
df_roll_test = pd.concat((y_roll_test, X_roll_test), axis=1)

df_roll_train.shape, df_roll_val.shape, df_roll_test.shape
```

```python
df_pitch_train = pd.concat((y_pitch_train, X_pitch_train), axis=1)
df_pitch_val = pd.concat((y_pitch_val, X_pitch_val), axis=1)
df_pitch_test = pd.concat((y_pitch_test, X_pitch_test), axis=1)

df_pitch_train.shape, df_pitch_val.shape, df_pitch_test.shape
```

## save

```python
df_roll_train.to_excel(data_folder + "6_roll__train.xlsx", index=False)
df_roll_val.to_excel(data_folder + "6_roll__val.xlsx", index=False)
df_roll_test.to_excel(data_folder + "6_roll__test.xlsx", index=False)

df_pitch_train.to_excel(data_folder + "6_pitch__train.xlsx", index=False)
df_pitch_val.to_excel(data_folder + "6_pitch__val.xlsx", index=False)
df_pitch_test.to_excel(data_folder + "6_pitch__test.xlsx", index=False)
```
