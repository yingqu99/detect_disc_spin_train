### hotmap

```python
df_corr_roll_raw.shape
```

```python
import plotly.figure_factory as ff
import plotly.express as px

mask = np.triu(np.ones_like(df_corr_roll_raw, dtype=bool))
df_mask = df_corr_roll_raw.mask(mask)

fig = ff.create_annotated_heatmap(z=df_mask.to_numpy(),
                                  x=df_mask.columns.tolist(),
                                  y=df_mask.columns.tolist(),
                                  colorscale=px.colors.diverging.RdBu,
                                  hoverinfo="none",  #Shows hoverinfo for null values
                                  showscale=True, ygap=1, xgap=1
                                  )

fig.update_xaxes(side="bottom")

fig.update_layout(
    title_text='Heatmap - roll',
    title_x=0.5,
    width=2000,
    height=2000,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    xaxis_zeroline=False,
    yaxis_zeroline=False,
    yaxis_autorange='reversed',
    template='plotly_white'
)

# NaN values are not handled automatically and are displayed in the figure
# So need to get rid of the text manually
for i in range(len(fig.layout.annotations)):
    if fig.layout.annotations[i].text == 'nan':
        fig.layout.annotations[i].text = ""

fig.show()
```

```python
df_high_corr_roll = []
threshold = 0.7

for col in df_corr_roll.columns:
    # Get correlations for this feature (excluding self-correlation)
    if col == 'pitch_true':
        continue
        
    correlations = abs(df_corr_roll.loc['roll_true', col])
    # Check if any correlation exceeds threshold
    if (correlations >= threshold).any():
        df_high_corr_roll.append(col)

# Create subset correlation matrix with only highly correlated features
df_high_corr_roll = df_corr_roll.loc[df_high_corr_roll, df_high_corr_roll]
```

```python
df_high_corr_roll.shape
```

```python
df_high_corr_roll
```

```python

mask = np.triu(np.ones_like(df_high_corr_roll, dtype=bool))
df_mask = df_high_corr_roll.mask(mask)

fig = ff.create_annotated_heatmap(z=df_mask.to_numpy(),
                                  x=df_mask.columns.tolist(),
                                  y=df_mask.columns.tolist(),
                                  colorscale=px.colors.diverging.RdBu,
                                  hoverinfo="none",  #Shows hoverinfo for null values
                                  showscale=True, ygap=1, xgap=1
                                  )

fig.update_xaxes(side="bottom")

fig.update_layout(
    title_text='Heatmap high corr - roll',
    title_x=0.5,
    width=800,
    height=800,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    xaxis_zeroline=False,
    yaxis_zeroline=False,
    yaxis_autorange='reversed',
    template='plotly_white'
)

# NaN values are not handled automatically and are displayed in the figure
# So need to get rid of the text manually
for i in range(len(fig.layout.annotations)):
    if fig.layout.annotations[i].text == 'nan':
        fig.layout.annotations[i].text = ""

fig.show()
```
